```python
import numpy as np
import cv2
import glob
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# prepare object points
# prepare object points, like (0,0,0), (1,0,0), (2,0,0) ....,(6,5,0)
objp = np.zeros((6*8,3), np.float32)
objp[:,:2] = np.mgrid[0:8, 0:6].T.reshape(-1,2)

#print(objp)

# Arrays to store object points and image points from all the images.
objpoints = [] # 3d points in real world space
imgpoints = [] # 2d points in image plane.

# Make a list of calibration images
images = glob.glob('camera_cal/*.jpg')

# Step through the list and search for chessboard corners
for idx, fname in enumerate(images):
    img = cv2.imread(fname)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Find the chessboard corners
    ret, corners = cv2.findChessboardCorners(gray, (8,6), None)

    # If found, add object points, image points
    if ret == True:
        objpoints.append(objp)
        imgpoints.append(corners)

        # Draw and display the corners
        cv2.drawChessboardCorners(img, (8,6), corners, ret)

        #write_name = 'corners_found'+str(idx)+'.jpg'
        #cv2.imwrite(write_name, img)
        
        plt.imshow(img)
```

```python
print(objpoints)
print(imgpoints)
```

```python
import pickle

# Test undistortion on an image
img = cv2.imread('test_images/test1.jpg')
img_size = (img.shape[1], img.shape[0])

# Do camera calibration given object points and image points
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, img_size, None, None)

dst = cv2.undistort(img, mtx, dist, None, mtx)
cv2.imwrite('output_images/test1_undist.jpg', dst)

# Save the camera calibration result for later use (we won't worry about rvecs / tvecs)
dist_pickle = {}
dist_pickle["mtx"] = mtx
dist_pickle["dist"] = dist
pickle.dump(dist_pickle, open("camera_cal/calibration_pickle.p", "wb"))

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
dst = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)

# Visualize undistortion
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.imshow(img)
ax1.set_title('Original Image', fontsize=30)

ax2.imshow(dst)
ax2.set_title('Undistorted Image', fontsize=30)
```

```python
import matplotlib.image as mpimg

def show_2_images(img1, img1_label, img2, img2_label, cmap = None):
    # Visualize undistortion
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

    ax1.imshow(img1, cmap=cmap)
    ax1.set_title(img1_label, fontsize=30)

    ax2.imshow(img2, cmap=cmap)
    ax2.set_title(img2_label, fontsize=30)


img = mpimg.imread('test_images/test1.jpg')
img2 = mpimg.imread('test_images/test3.jpg')

show_2_images(img, 'test1', img2, 'test3')
```

```python
kernel = 9
grad_x_binary = abs_sobel_thresh(img, orient='x', sobel_kernel=kernel, thresh=(20, 100))

grad_x_binary2 = abs_sobel_thresh(img2, orient='x', sobel_kernel=kernel, thresh=(20, 100))

show_2_images(grad_x_binary, 'test1', grad_x_binary2, 'test3', cmap='gray')
```

```python
kernel = 9
grad_y_binary = abs_sobel_thresh(img, orient='y', sobel_kernel=kernel, thresh=(20, 100))
grad_y_binary2 = abs_sobel_thresh(img2, orient='y', sobel_kernel=kernel, thresh=(20, 100))

show_2_images(grad_y_binary, 'test1', grad_y_binary2, 'test3', cmap='gray')
```

```python
kernel = 9
grad_mag_binary = mag_thresh(img, sobel_kernel=kernel, mag_thresh=(30, 100))
grad_mag_binary2 = mag_thresh(img2, sobel_kernel=kernel, mag_thresh=(30, 100))

show_2_images(grad_mag_binary, 'test1', grad_mag_binary2, 'test3', cmap='gray')
```

```python
kernel = 9
grad_dir_binary = dir_threshold(img, sobel_kernel=kernel, thresh=(0.7, 1.3))
grad_dir_binary2 = dir_threshold(img2, sobel_kernel=kernel, thresh=(0.7, 1.3))
show_2_images(grad_dir_binary, 'test1', grad_dir_binary2, 'test3', cmap='gray')
```

```python
combined = np.zeros_like(grad_dir_binary)
combined[((grad_x_binary == 1) & (grad_y_binary == 1)) | ((grad_mag_binary == 1) & (grad_dir_binary == 1))] = 1

combined2 = np.zeros_like(grad_dir_binary2)
combined2[((grad_x_binary2 == 1) & (grad_y_binary2 == 1)) | ((grad_mag_binary2 == 1) & (grad_dir_binary2 == 1))] = 1

show_2_images(combined, 'test1', combined2, 'test3', cmap='gray')
```
